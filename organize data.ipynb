# combine movielens and imdb movies to create movie table 

```python
movielens_movies = pd.read_csv('dataset/movielens_movies.csv')
```

need to split the title in movie_lens to 'title' and 'year' column 

```python
import re
```

```python
title_split = movielens_movies.title.str.split('\(|\)',expand=True,n=2)
```

```python
movielens_movies.title =title_split[0].str.strip()
```

```python
movielens_movies['year']=title_split[1]
```

```python
movielens_movies.head()
```

```python
imdb_movies = pd.read_csv('dataset/IMDB_movies.csv')
```

```python
imdb_movies.head()
```

we only want certain columns 

```python
imdb_movies.columns
```

```python
imdb_movies = imdb_movies[['imdb_title_id','original_title','year','genre','duration','description']]
```

```python
imdb_movies.rename(columns={'original_title':'title'}, inplace=True)
```

```python
imdb_movies.head()
```

## inner join movielens_movies and imdb_movies on title

```python
movie_join = movielens_movies.merge(imdb_movies, on='title', how='inner')
movie_join.head()
```

```python
movies_table = movie_join[['movieId','title','year_x','imdb_title_id','duration','description']]
movies_table.rename(columns={'movieId':'movielens_title_id','year_x':'year'}, inplace=True)
movies_table
```

## calculate ratings from ratings_table 

```python
ratings_table.head()
```

```python
num_ratings = ratings_table.groupby(['movieId'], as_index=False).count()[['movieId','rating']]
num_ratings.rename(columns={'rating':'num_ratings'}, inplace=True)
num_ratings.head()
```

```python
ave_ratings = ratings_table.groupby(['movieId'], as_index=False).mean()[['movieId','rating']]
ave_ratings.rename(columns={'rating':'ave_rating'}, inplace=True)
ave_ratings.head()
```

```python
movies_table.shape
```

```python
# use only movie with ratings 
tmp=pd.merge(movies_table, num_ratings,how='inner', left_on='movielens_title_id',right_on='movieId')
tmp=pd.merge(tmp, ave_ratings,how='inner', left_on='movielens_title_id',right_on='movieId')
tmp.head()
```

```python
movies_table = tmp 
```

#  actors table

```python
actors_table = imdb_names[['imdb_name_id','name','bio']]
actors_table.head()
```

# casts table 

```python
imdb_casts = pd.read_csv('dataset/IMDb_title_principals.csv')
imdb_casts.head()
```

```python
# only want movies that are in our movies table 
imdb_casts = imdb_casts[imdb_casts.imdb_title_id.isin(movies_table.imdb_title_id)] 
# select only actor and actress 
imdb_casts= imdb_casts[imdb_casts.category.isin(['actor','actress'])] 
```

```python
casts_table = imdb_casts[['imdb_title_id','imdb_name_id']]
```

```python
casts_table.head()
```

# genre table 

```python
movie_join.head()
```

```python
# use movielens's genres column 
genres = movie_join[['imdb_title_id','genres']][movie_join.imdb_title_id.isin(movies_table.imdb_title_id)]
genres.head()
```

```python
genres
```

```python
genres_table = genres[['imdb_title_id']].join(\
    genres.genres.str.split('|',expand=True)\
    .stack().reset_index(drop=True, level=1).rename('genres'))
genres_table.head()
```
